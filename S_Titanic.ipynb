### Impute null values

### We use mode for categorical columns because it is best approach to fill null values with mode and use indexing to fill with first category in the each categorical columns

```python
def fill_with_mode(columns):
  for col in columns:
    train_df[col] = train_df[col].fillna(train_df[col].mode()[0])
  return
  
fill_with_mode(columns=["HomePlanet", "CryoSleep", "Destination", "VIP"])
  
```

```python
def fill_with_mode(columns):
  for col in columns:
    test_df[col] = test_df[col].fillna(test_df[col].mode()[0])
  return
  
fill_with_mode(columns=["HomePlanet", "CryoSleep", "Destination", "VIP"])
  
```

### We use median here becaues in numeric columns data is skewed and contain big outliers so fill with median is best approach 

```python
def fill_with_median(columns):
  for col in columns:
    train_df[col] = train_df[col].fillna(train_df[col].median())
  return
  
fill_with_median(columns=["Age","VIP", "RoomService", "FoodCourt", "ShoppingMall","Spa","VRDeck"])
  
```

# **Normalizing**

```python
train_df.head(8)
```

```python
scaler = MinMaxScaler(feature_range=(0,1))
```

### We have to normalize the data here because the values are not normally distributed there. To take them on same scale for better traing and for good accuracy we use **`Minimax scaler()`**

```python
def minimaxscaler_for_training_data(columns):
  for col in columns:
    train_df[[col]] = scaler.fit_transform(train_df[[col]])
    

minimaxscaler_for_training_data(columns=["Age","RoomService","FoodCourt","ShoppingMall","Spa","VRDeck"])
```

```python
def minimaxscaler_for_testing_data(columns):
  for col in columns:
    test_df[[col]] = scaler.fit_transform(test_df[[col]])
    

minimaxscaler_for_testing_data(columns=["Age","RoomService","FoodCourt","ShoppingMall","Spa","VRDeck"])
```

### **Train and Test Split**

```python
X = train_df.drop(columns="Transported")
Y = train_df["Transported"]

X_test = test_df.copy()
```

```python
X_train_part, X_values, Y_train_part, Y_values = train_test_split(X, Y, test_size=0.2, random_state=42)
```

### **Training**

```python
model = RandomForestClassifier()
model.fit(X_train_part, Y_train_part)
```

### **Model Evaluation**

```python
y_predict = model.predict(X_values)
print(f"Accuuracy Score: {accuracy_score(Y_values,y_predict)*100:.2f}")
```

```python
test_predictions = model.predict(X_test)
```
